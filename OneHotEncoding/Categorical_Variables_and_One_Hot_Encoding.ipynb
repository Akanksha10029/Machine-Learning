#Dummy Variable Trap
When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

**NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you**

```python
final = final.drop(['west windsor'],axis=1)
final
```

```python
reg = linear_model.LinearRegression()
reg.fit(final.drop('price',axis='columns'),df.price)
```

#Using sklearn OneHotEncoder

**In order to use OneHotEncoder we need to use it on town column so, for this we are going use LabelEncoder from sklearn preprocessing**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

**fit_transform means it takes the label column as an input so, here we have label column ,i.e, town  and it will return the lavels in array form...**

**First step is to use label encoder to convert town names into numbers**

```python
dfle = df
le.fit_transform(dfle.town)
```

#or

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
x = dfle[['town','area']].values
x
```

```python
y = dfle.price
y
```

**Now use one hot encoder to create dummy variables for each of the town**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

**[0] here it is specifying that whatever x i am supplying , the 0th column in that x is my categorical feature**

```python
df.head(10)
```

```python
x = ct.fit_transform(x)
x
```

**To avoid dummy variable trap I am going to drop one of the column so, here i am droping 1st column.**

```python
x = x[:,1:]
```

In [:,1.:] I am taking all the rows and for the columns take all the columns from index 1 onwards...This will essentially drop our 0th column

```python
x
```

```python
model.fit(x,y)
```
