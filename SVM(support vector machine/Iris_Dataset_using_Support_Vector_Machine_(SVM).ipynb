#Box Plot
**Median (Middle Line):** The middle line within the box represents the median (or the 50th percentile) of the dataset. It's the value where half of the data falls below it, and half falls above it.

**Box (Quartiles):** The box represents the interquartile range (IQR), which spans from the first quartile (Q1) to the third quartile (Q3). Q1 represents the 25th percentile, and Q3 represents the 75th percentile. The box itself contains 50% of the data. The width of the box displays the spread or variability within this middle 50% of the dataset.

**Whiskers:** The lines (whiskers) extend from the box to the minimum and maximum values that are within a calculated range, usually 1.5 times the IQR from the upper and lower quartiles. Any data points beyond this range are considered outliers and plotted as individual points.

**Outliers:**Individual points or dots beyond the whiskers are potential outliers in the dataset, which lie significantly farther from the median.

#Here for setosa flower we can see that the box is small, so what does it means?
**A small box implies that the middle 50% of the data points (contained within the box) are concentrated in a narrower range of values.
It suggests that the data points are closer together within this range, showing less variability or spread in that part of the dataset.**

**A big box implies high variability or spread in that part of the dataset**

```python
from sklearn.model_selection import train_test_split
```

```python
df.head()
```

```python
x= df.drop(['Flower_name','species'],axis=1)
x
```

```python
y=df['species']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix

#Make predictions on the test set
y_pred = model.predict(x_test)

# Create a confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
import seaborn as sns
```

```python
# Define class names
class_names = ['setosa', 'versicolor', 'virginica']

# Create a heatmap for the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Greens', fmt='d', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
```
