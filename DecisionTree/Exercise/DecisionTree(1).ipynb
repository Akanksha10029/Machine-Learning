#Decision Tree Classification

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/salaries(DecisionTree).csv')
df
```

```python
inputs = df.drop('salary_more_then_100k',axis=1)
inputs
```

```python
target = df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['Company']=le_company.fit_transform(inputs['company'])
inputs['Job']=le_job.fit_transform(inputs['job'])
inputs['Degree']=le_degree.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
Inputs = inputs.drop(['company','job','degree'],axis=1)
Inputs
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(Inputs, target)
```

```python
model.score(Inputs,target)
```

```python
model.predict([[2,1,0]])
```
