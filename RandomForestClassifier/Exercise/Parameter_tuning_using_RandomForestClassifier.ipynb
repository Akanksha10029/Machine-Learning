```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/Iris.csv')
df
```

```python
df.drop('Id', axis=1, inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
species_n=LabelEncoder()
df['species']=species_n.fit_transform(df['Species'])
df
```

```python
df = df.rename(columns={'Species': 'Flower_name'})
df
```

```python
x= df.drop(['Flower_name','species'],axis=1)
x
```

```python
y=df.species
y
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

# range of n_estimators to explore
grid = {'n_estimators': [10, 50, 100, 150, 200]}

# Performing Grid Search to find the best 'n_estimators'
grid_search = GridSearchCV(estimator=rf, param_grid=grid, cv=4)
grid_search.fit(x, y)

# Print the best parameters and best score
print("Best n_estimators:", grid_search.best_params_['n_estimators'])
print("Best Score:", grid_search.best_score_)
```

**Parameter tuning using RandomForestClassifier**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

# range of n_estimators to explore
grid = {'n_estimators': [10, 50, 100, 150, 200,250]}

# Performing Grid Search to find the best 'n_estimators'
grid_search = GridSearchCV(estimator=rf, param_grid=grid, cv=4)
grid_search.fit(x, y)

# Print the best parameters and best score
print("Best n_estimators:", grid_search.best_params_['n_estimators'])
print("Best Score:", grid_search.best_score_)
```
