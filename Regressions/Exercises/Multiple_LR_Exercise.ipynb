**This file contains** hiring statics for a firm such as **experience of candidate, his written test score and personal interview score.** Based on these 3 factors, HR will decide the salary. Given this data, you need to **build a machine learning model for HR department that can help them decide salaries for future candidates.** Using this predict salaries for following candidates,

2 yr experience, 9 test score, 6 interview score

12 yr experience, 10 test score, 10 interview score

Answer

1.   53713.86
2.   93747.79

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
!pip install word2number
from word2number import w2n
```

```python
df = pd.read_csv('/content/hiring(multiple linear reg).csv')
df
```

```python
df.experience = df.experience.fillna("zero")
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
mean_score=df['test_score(out of 10)'].mean()
mean_score_int = int(mean_score)
mean_score_int
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(mean_score_int)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

#OR(we can train our model in this way also by dropping the dependent variable i.e salary($))

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('salary($)',axis='columns'),df['salary($)'])
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
