#Naive Bayes: Survival Prediction from titanic crash

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/titanic.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:20]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean()).astype(int)
inputs.head(10)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(10)
```

**Dropping male column as well because of dummy variable trap theory. One column is enough to repressent male vs female**

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
x_test[0:10]
```

```python
y_test[0:10]
```

```python
model.predict(x_test)
```

```python
model.predict_proba(x_test[:10])
```

**Here above we can see that first column is depicting "survived" and second column is depicting "NOT survived"**

**Just match predict_proba column values with the above model.predict values like first is 0 i.e NOT survived then in second column we can see that probability is less than 5 therefore second column is showing not survived**

**Now Calculate the score using cross validation**

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),x_train, y_train, cv=5)
```
